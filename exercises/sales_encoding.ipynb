## Encode division (Label)

```python
# TODO add an LabelEncoder() for division
le = LabelEncoder()
df['division_enc'] = le.fit_transform(df['division'])
print(le.classes_)
print(df[['division', 'division_enc']].head(10))
```

## Ordinal Encoding of level of education

```python
# TODO add an ordinal encoding of level of education
education_order = ["associate's degree", 'some college', 'high school', "bachelor's degree", "master's degree"]
ordinal_encoder = OrdinalEncoder(categories=[education_order])
df['education_level_enc'] = ordinal_encoder.fit_transform(df[['level of education']])
print(df[['level of education', 'education_level_enc']].head(10))
```

## One Hot Encoding (division)

```python
# TODO add a one hot encoding of division https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html
from sklearn import preprocessing
enc = preprocessing.OneHotEncoder()
encoded_division = enc.fit_transform(df[['division']]).toarray()
encoded_division_df = pd.DataFrame(encoded_division, columns=enc.get_feature_names_out(['division']))
df = pd.concat([df, encoded_division_df], axis=1)
print(df.head(10))
```
